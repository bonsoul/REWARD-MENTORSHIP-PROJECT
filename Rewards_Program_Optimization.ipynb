```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/Data Sets Reward_Program_Assignment_Input_v6 - TA.xlsx - Mentorship_Sessions.csv')
```

```python
data.drop(['Unnamed: 0'], axis=1, inplace=True) # Change axis to 1 to drop a column
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# Drop rows with null values in 'Mentor_Name'
data.dropna(subset=['Mentor_Name'], inplace=True)

# Replace null values in 'points_awarded' with 0
data['Points_Awarded'].fillna(0, inplace=True)
```

```python
data.head()
```

```python
# Get a list of all columns except 'points_awarded'
columns_to_consider = data.columns.difference(['Points_Awarded'])

# Drop rows with null values in the specified columns
data.dropna(subset=columns_to_consider, inplace=True)

# Replace null values in 'points_awarded' with 0
data['Points_Awarded'].fillna(0, inplace=True)
```

```python
data.isnull().sum()
```

```python
# Correct 'Job_Info_Completed' column by replacing inconsistent values
data['Job_Info_Completed'] = data['Job_Info_Completed'].replace({'Maybe': 'No'})
```

```python
#check duplicates

data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
# Helper function to calculate points for a single row
def calculate_points(row):
    Session_Duration = row['Session_Duration_Min']
    Job_Info_Completed = row['Job_Info_Completed']

    # Initialize points
    points = 250  # Base points for signing up

    # Award points for session duration
    if Session_Duration >= 30:
        points += 250

    # Limit points for session duration to 500
    points = min(points, 500)

    # Award points for job info completion
    if Job_Info_Completed == 'Yes':
        points += 250  # Assuming you want to award 250 for job info

    return points

# Apply the function to create the 'Points_Awarded' column
data['Points_Awarded'] = data.apply(calculate_points, axis=1)

# Award 1000 points if the mentor has mentored 2 different mentees
# We'll need to adjust this part based on your desired logic
# for awarding points for mentoring multiple mentees.

# Example: Awarding 1000 points to mentors with 2+ different mentees
for mentor_id in data['Mentor_ID'].unique():
    if data[data['Mentor_ID'] == mentor_id]['Mentee_Name'].nunique() >= 2:
        data.loc[data['Mentor_ID'] == mentor_id, 'Points_Awarded'] += 1000

# Display the DataFrame with the Points_Awarded column
data.sort_values(by='Points_Awarded', ascending=False)
data.head()
```

```python
# prompt: Save the final dataset into csv filee

data.to_csv('final_dataset.csv', index=False)
```
